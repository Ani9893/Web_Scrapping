# Single Car Company Scraping

```python
import requests
from bs4 import BeautifulSoup as bs

h = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36'}

id = requests.get('https://www.cardekho.com/maruti-suzuki-cars',headers = h)
id
```

```python
id.status_code
```

```python
id.text
```

```python
so = bs(id.text)
so
```

```python
title = so.find('h1').text
title
```

```python
car_name = []
for i in so.find_all('h3'):                 # car name
  car_name.append(i.text)
car_name = car_name[:23]
car_name
```

```python
price = []
for i in so.find_all('div',class_='price'):        # price
  price.append(i.text)
price = price[:23]
price
```

```python
milage = []
for i in so.find_all('div',class_='dotlist'):     # milage
  milage.append(i.text)
milage = milage[:23]
milage
for j in milage:
  print(j.split())
```

```python
type_li = []
mileage_li = []
transmission_li = []

for i in so.find_all('div',class_='dotlist')[:23]:
  li = []
  for j in (i.find_all('span')):
    val = j.text
    li.append(val)
  if '.' not in li[1]:
    li.insert(1,None)
  a,b,c = li
  type_li.append(a)
  mileage_li.append(b)
  transmission_li.append(c)

print(len(type_li), len(mileage_li), len(transmission_li))

print(type_li)
print(mileage_li)
print(transmission_li)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['Car_name'] = car_name
df['Price'] = price
df['Mileage'] = mileage_li
df['Type'] = type_li
df['Transmission'] = transmission_li
df
```

# All Company Car Scraping

```python
j = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36'}

id = requests.get('https://www.cardekho.com/newcars#brands',headers=j)
id
```

```python
id.text
```

```python
mm = bs(id.text)
mm
```

```python
s = mm.find_all('a',class_='BrIconNewCar')
s[0]['href']
```

```python
links =[]
for i in s:
  links.append('https://www.cardekho.com' + i['href'])
links
```

```python
import requests
def get_details():

  jk = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36'}

  n,p,m,l = [],[],[],[]
  for link in links:
    s = bs(requests.get(link,headers=jk).text)
    for j in s.find_all('div',class_="gsc_col-sm-12 gsc_col-xs-12 gsc_col-md-8 listView holder posS"):
      n.append(j.find("h3").text)
      p.append(j.find("div",class_="price").text)
      m.append(j.find("div",class_="dotlist").text)
      l.append(j.find('a',title = True)['href'].split('/')[-2])

  d = pd.DataFrame({"Name":n,"Price":p,"Milage":m,'brand':l})
  return d
get_details()
```

```python
df = get_details()
df.tail()
```

```python
df.to_csv('car_details.csv')
```
